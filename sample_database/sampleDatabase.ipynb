```python
smallerSchoolStudent = studentData[['studentID', 'schoolID', 'year']]

smallerSchoolStudent = smallerSchoolStudent.rename(columns={"year": "yearAttended"})

smallerSchoolStudent
```

```python
teacherData.rename(columns={'grade':'gradeLevel','teacherID': 'staffID'}, inplace= True)
teacherData['position'] = 'teacher'
teacherData.head(2)
```

```python
smallerStaff = teacherData[['staffID', 'name', 'gradeLevel', 'position']]

smallerStaff
```

```python
# staffStudent

# trim teacherData
dfteacher = teacherData[['name','middleSchool', 'staffID', 'course']]
dfteacher['courseID'] = [1 if ele  == "Language Arts" else 2 if ele == 'Social Studies' else 3 if ele == 'Health/Physical Education' else 4 if ele == 'Mathematics' else 5 if ele == 'Science' else 6 if ele == 'Life Science' else 7 for ele in dfteacher.course]


#trim studentData
dfstudent = studentData[['teacher','year','school', 'studentID','schoolID']]
# merge trimmed dataframes
merged_data = pd.merge(dfteacher, dfstudent, "inner", left_on=['name','middleSchool'], 
                       right_on = ['teacher','school']).drop(['middleSchool', 'name'], axis=1)
merged_data
```

```python
smallerStaffStudent = merged_data[['staffID', 'studentID', 'year']]
smallerStaffStudent.drop_duplicates()
```

```python
# staffCourses

# Don't have course ID
# ['staffID', 'courseID', 'year'])

smallerStaffCourses = merged_data[['staffID', 'courseID', 'year']]

smallerStaffCourses.drop_duplicates()
```
